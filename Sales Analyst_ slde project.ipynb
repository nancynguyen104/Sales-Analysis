```python
import pandas as pd  
```

```python

df_sales= pd.read_csv("C:\\Users\\61410\\Downloads\\Data Wrangling Practise.zip (1)\\Data Wrangling Practise\\sales.csv")
print(df_sales)
```

```python

df_prices = pd.read_csv("C:\\Users\\61410\\Downloads\\Data Wrangling Practise.zip (1)\\Data Wrangling Practise\\prices.csv")
print(df_prices)
```

```python
#clean data
#Convert datetime
df_sales['ordered_at'] = pd.to_datetime(df_sales['ordered_at'])
df_sales['ordered_at'] = df_sales['ordered_at'].dt.strftime('%Y-%m-%d %H:%M:%S')
df_prices['updated_at'] = pd.to_datetime(df_prices['updated_at'])
```

```python
#Rank update at in price table
df_prices = df_prices.sort_values(by =['product_id','updated_at'],ascending = [True,True])
print(df_prices)
```

```python
#Join tables 

df_join = pd.merge(df_sales,df_prices, on = 'product_id', how = 'left')

#remove rows where order_at before update_at
# case 1: if order_at > update_at
filter_before = df_join[df_join['ordered_at'] >= df_join['updated_at']]
filter_before
```

```python
#group by product_id

filter_before['rank_updated'] = filter_before.groupby(['product_id','ordered_at'])['updated_at'].rank(method ='dense',ascending=False)
filter_before
```

```python
filter_before = filter_before[filter_before['rank_updated']== 1.0]
filter_before = filter_before.drop_duplicates()
filter_before
```

```python
#join back with sales 
df_final_price = pd.merge(df_sales,filter_before, on = ['product_id','ordered_at'], how = 'left')
df_final_price
```

```python
#total_order
summary = df_final_price.groupby(['product_id','new_price'])['quantity_ordered_y'].sum().reset_index(name = "total_order")
summary.rename(columns={"new_price" : "final_price"},inplace= True)
summary
```

```python
#if order at < update_at 
df_prices['rank_update'] = df_prices.groupby('product_id')['updated_at'].rank(method = 'dense', ascending= True)
df_prices_oldest_update = df_prices[df_prices['rank_update'] == 1]
df_prices_oldest_update
```

```python
df_old_order = pd.merge(df_prices_oldest_update,df_sales, on = ['product_id'], how = 'right')
df_old_order = df_old_order[df_old_order['ordered_at'] < df_old_order['updated_at']]
df_old_order
```

```python
summary1 = df_old_order.groupby(['product_id','old_price'])['quantity_ordered'].sum().reset_index(name = "total_order")
summary1.rename(columns={"old_price" : "final_price"},inplace= True)
summary1
```

```python
#union
all_sales = pd.concat([summary,summary1])
all_sales
```

```python
#revenue

all_sales['revenue'] = all_sales['final_price'] * all_sales['total_order']
all_sales
```

```python
#total_revenue 

summary_revenue = all_sales.groupby(['product_id'])['revenue'].sum()
summary_revenue
```
